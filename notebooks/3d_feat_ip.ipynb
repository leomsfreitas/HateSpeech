```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

from src.utils.features import load_features

from src.utils.path import DATA_PATH
```

```python
FEATURE_COLS = ["feat_pos_ratio", "feat_neg_ratio", "feat_nto_ratio"]
```

# Feature Extraction Functions

```python
def compute_ip(df: pd.DataFrame, feature_columns: list[str], label_column: str):
    X = df[feature_columns]
    y = df[label_column]

    model = LogisticRegression(class_weight="balanced")
    model.fit(X, y)

    coef = pd.Series(model.coef_[0], index=feature_columns)
    ip = model.intercept_[0] + sum(coef[col] * df[col] for col in feature_columns)

    return pd.Series(ip, index=df.index, name="ip")
```

# Feature Engineering in Corpora

## Hate-BR

```python
hate_lex = load_features(
    base_path=f"{DATA_PATH}/hate-br/processed/hate-br.csv",
    feature_paths=[f"{DATA_PATH}/hate-br/features/lex.csv"]
)
```

```python
hate_ip = compute_ip(hate_lex, FEATURE_COLS, "offensive_language")
```

```python
hate_ip.to_csv(f"{DATA_PATH}/hate-br/features/ip.csv", index=False)
```
