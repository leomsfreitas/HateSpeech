# Binary Classification

## Load Base Data

```python
df_base = pd.read_csv(f"{DATA_PATH}/hate-br/processed/hate-br.csv")
```

## 1. Baseline

```python
config_baseline = AutoGluonConfig(
    model_path="./models/automl/binary/baseline",
    label_columns=["offensive"],
    eval_metrics=["f1_macro"]
)
```

```python
pipeline_baseline = PipelineAutoGluonBinary(config_baseline, df_base)
train, val, test = pipeline_baseline.split(df_base)
pipeline_baseline.train(train, val)
```

```python
y_true, y_pred = pipeline_baseline.evaluate(test)
print(classification_report(y_true, y_pred))
```
