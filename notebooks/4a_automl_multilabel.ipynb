# Multilabel Classification

```python
LABEL_COLUMNS = ["insult", "obscene", "ideology", "lgbtqphobia", "racism", "sexism", "xenophobia"]
```

## Load Base Data

```python
df_base = pd.read_csv(f"{DATA_PATH}/unified-hate/processed/unified-hate.csv")
```

## 1. Baseline Multilabel

```python
config_baseline = AutoGluonConfig(
    model_path="./models/automl/multilabel/baseline",
    label_columns=LABEL_COLUMNS,
    eval_metrics=["f1_macro"] * len(LABEL_COLUMNS)
)
```

```python
pipeline_baseline = PipelineAutoGluonMultilabel(config_baseline, df_base)
train, val, test = pipeline_baseline.split(df_base)
pipeline_baseline.train(train, val)
```

```python
y_true, y_pred = pipeline_baseline.evaluate(test)
multilabel_report(y_true, y_pred)
```
