```python
import pandas as pd

from src.utils.path import DATA_PATH
```

# Load Data

```python
hate = pd.read_csv(f"{DATA_PATH}/hate-br/processed/hate-br.csv")
olid = pd.read_csv(f"{DATA_PATH}/olid-br/processed/olid-br.csv")
told = pd.read_csv(f"{DATA_PATH}/told-br/processed/told-br.csv")
```

## Selected Classes

```python
classes = ["text", "insult", "obscene", "ideology", "lgbtqphobia", "racism", "sexism", "xenophobia"]
```

# Adapting Corpora

```python
olid = olid.rename(columns={"profanity_obscene": "obscene"})
olid = olid.reindex(columns=classes, fill_value=0)
```

```python
told = told.rename(columns={"homophobia": "lgbtqphobia", "misogyny":"sexism"})
told = told.reindex(columns=classes, fill_value=0)
```

```python
hate = hate.rename(columns={"instagram_comments": "text", "partyism": "ideology", "homophobia": "lgbtqphobia"})
hate = hate.reindex(columns=classes, fill_value=0)
```

# Concat Corpora

```python
unified = pd.concat([olid, told, hate], ignore_index=True)
```

```python
unified.drop_duplicates(subset=["text"], inplace=True)
```

```python
unified = unified[unified.eq(1).any(axis=1)]
```

```python
unified.dropna(inplace=True)
```

```python
unified.to_csv(f"{DATA_PATH}/unified-hate/processed/unified-hate.csv", index=False)
```
