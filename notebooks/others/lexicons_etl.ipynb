# LIWC

Tags 126 (Positivo) e 127 (Negativo) determinam a polaridade de cada palavra.

```python
with open(f"{RAW_LEXICONS_PATH}/LIWC.txt", "r") as f:
    lines = f.readlines()

liwc = {}
for line in lines:
    parts = line.strip().split()
    if not parts or "*" in parts[0]:
        continue
    word = parts[0]
    tags = list(map(int, parts[1:]))
    if 126 in tags and 127 in tags:
        pol = 0
    elif 126 in tags:
        pol = 1
    elif 127 in tags:
        pol = -1
    else:
        pol = 0
    liwc[word] = pol
```

```python
df_liwc = pd.DataFrame(list(liwc.items())[1:], columns=["word", "polarity"])
df_liwc.to_csv(f"{PROCESSED_LEXICONS_PATH}/liwc.csv", index=False)
```

# SentiLex

`POL:N0` → polaridade quando alvo humano é sujeito.
`POL:N1` → polaridade quando alvo humano é complemento.

```python
with open(f"{RAW_LEXICONS_PATH}/SentiLex-lem-PT02.txt", "r") as f:
    lines = f.readlines()

sentilex = {}
for line in lines:
    line = line.replace("\n", "").split(";")
    word = line[0].split(".")[0]
    polarity = line[2].split("=")[1]
    sentilex[word] = polarity
```

```python
df_sentilex = pd.DataFrame(list(sentilex.items())[1:], columns=["word", "polarity"])
df_sentilex.to_csv(f"{PROCESSED_LEXICONS_PATH}/sentilex.csv", index=False)
```

# WordNetAffect

```python
with open(f"{RAW_LEXICONS_PATH}/wordnetaffectbr_valencia.csv", "r", encoding="latin-1") as f:
    lines = f.read().splitlines()

wordnetaffect = {}
for line in lines:
    word, polarity = line.split(";")
    wordnetaffect[word] = -1 if polarity == "-" else 1
```

```python
df_wordnetaffect = pd.DataFrame(list(wordnetaffect.items())[1:], columns=["word", "polarity"])
df_wordnetaffect.to_csv(f"{PROCESSED_LEXICONS_PATH}/wordnetaffect.csv", index=False)
```
