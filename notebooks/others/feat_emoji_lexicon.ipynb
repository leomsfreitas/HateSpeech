## Load Emoji Lexicon

```python
emoji_lex = pd.read_csv(f"{PROCESSED_LEXICONS_PATH}/emoji_lex.csv")
emoji_polarity_map = emoji_lex.set_index("Emoji")["polarity"]
```

# Feature Extraction Functions

```python
def extract_emojis(text):
    found_emojis = []
    sorted_emojis = sorted(emoji_polarity_map.keys(), key=len, reverse=True)
    for emoji in sorted_emojis:
        if emoji in str(text):
            found_emojis.append(emoji)
    return found_emojis

def get_final_polarity(emoji_list):
    polarities = [emoji_polarity_map.get(emoji) for emoji in emoji_list if emoji_polarity_map.get(emoji) is not None]
    total = sum(polarities) if polarities else 0
    if total > 0: return 1
    if total < 0: return -1
    return 0
```

# Feature Engineering in Corpora

## Hate-BR

```python
hate = pd.read_csv(f"{PROCESSED_DATA_PATH}/hate-br.csv")
```

```python
hate["emojis"] = hate["text"].apply(extract_emojis)
hate["emoji_polarity"] = hate["emojis"].apply(get_final_polarity)
```

```python
hate[["emoji_polarity"]].to_csv(f"{FEATURES_PATH}/hate-br-emoji.csv", index=False)
```

## Ited-BR

```python
ited = pd.read_csv(f"{PROCESSED_DATA_PATH}/ited-br.csv")
```

```python
ited["emojis"] = ited["text"].apply(extract_emojis)
ited["emoji_polarity"] = ited["emojis"].apply(get_final_polarity)
```

```python
ited[["emoji_polarity"]].to_csv(f"{FEATURES_PATH}/ited-br-emoji.csv", index=False)
```
