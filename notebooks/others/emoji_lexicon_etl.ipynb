## Load Emoji Sentiment Data

```python
emoji_df = pd.read_csv(f"{RAW_LEXICONS_PATH}/Emoji_Sentiment_Data_v1.0.csv",)
emoji_df.drop(columns=["Position", "Unicode name", "Unicode block"], inplace=True)
```

# Probability Calculation

$$p_{-} = \frac{N_{-}}{N_{-} + N_{0} + N_{+}}$$
$$p_{0} = \frac{N_{0}}{N_{-} + N_{0} + N_{+}}$$
$$p_{+} = \frac{N_{+}}{N_{-} + N_{0} + N_{+}}$$

```python
emoji_df["p-"] = emoji_df["Negative"] / emoji_df["Occurrences"]
emoji_df["p0"] = emoji_df["Neutral"] / emoji_df["Occurrences"]
emoji_df["p+"] = emoji_df["Positive"] / emoji_df["Occurrences"]
```

# Sentiment Score

```python
emoji_df["score"] = emoji_df["p+"] - emoji_df["p-"]
```

# Classe Dominante (ArgMax)

```python
polarity_mapping = {"p-": -1, "p0": 0, "p+": 1}
emoji_df["polarity"] = emoji_df[["p-", "p0", "p+"]].idxmax(axis=1).map(polarity_mapping)
```

# Save
