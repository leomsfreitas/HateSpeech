# Preprocessing

## Hate-BR

```python
hate = pd.read_csv(f"{DATA_PATH}/hate-br/raw/hate-br.csv")
```

```python
hate.iloc[54:59][["instagram_comments"]]
```

```python
hate.info()
```

```python
hate = hate.rename(columns={"offensive_language": "offensive", "instagram_comments": "text"})
```

```python
hate.drop(columns=["offensiveness_levels", "apology_for_the_dictatorship", "antisemitism",
                   "fatphobia", "offensive_&_non-hate_speech", "non-offensive", "religious_intolerance",
                   "specialist_1_hate_speech", "specialist_2_hate_speech", "specialist_3_hate_speech"],
          inplace=True)
```

```python
hate = preprocess_text(df=hate, text_col="text", abbrev_dict=dict_offensive)
```

```python
hate.drop_duplicates(subset="text", inplace=True)
hate.dropna(inplace=True)
```

```python
hate[hate.columns[1:]] = hate[hate.columns[1:]].astype(int)
```

```python
hate.info()
```

```python
hate.to_csv(f"{DATA_PATH}/hate-br/processed/hate-br.csv", index=False)
```
